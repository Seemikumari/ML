```python
#different ways of treating an outliers
#1. deletion
# the entire row containing an outlier will be removed
# data size is reduced
# the valuable information in the process is lost especially if dataset is small
# it is possible and would not cause any significant difference to the final model in case of large datasets.
#2.capping or imputing
# outliers is not removed but is replaced with the average value or the median value or the mode value whichever
# makes the least impact to the variable
#3.data transformation 
# the variable is transformed to its log value or the square or cube root since log or square /cube root of an 
# outlier is likely to be in range of log or square/cube roots of rest of the data points. 
#4.binning
#different bins are formed based on the values of the variables to treat the outlier
```

```python
import pandas as pd
data=pd.read_csv("Sample Data.csv")
```

```python
data
```

```python
q1=data["Sale Price"].quantile(0.25)
q3=data["Sale Price"].quantile(0.75)
```

```python
iqr=q3-q1
```

```python
upper_limit=q3+1.5*iqr
lower_limit=q1-1.5*iqr
upper_limit,lower_limit
```

```python
def limit_imputer(value):
    if value >upper_limit:
        return upper_limit
    if value< lower_limit:
        return lower_limit
    else:
        return value
```

```python
data["Sale Price"]=data["Sale Price"].apply(limit_imputer)
```

```python
data.dropna(inplace=True,axis=0,subset=['Sale Price'])
```
